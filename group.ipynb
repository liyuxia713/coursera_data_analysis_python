```python
def inst_cover(df, mask, cols):
    n = len(cols)
    a = df.groupby(cols[:n-1])[cols[-1]].nunique()
    b = df.groupby(cols[:n-2])[cols[-1]].nunique()
    c = a.to_frame().unstack()/b.values
    c.columns = c.columns.droplevel(0)
    return c#.reset_index()
    
cols = ['os', 'brand', 'app', 'uid']
c=inst_cover(df, mask, cols)
```
